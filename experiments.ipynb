```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
data=pd.read_csv('Churn_Modelling.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.describe()
```

```python
data=data.drop(['RowNumber',	'CustomerId'],axis=1)
```

```python
data.head()
```

```python
labelencoder=LabelEncoder()
data['Gender']=labelencoder.fit_transform(data['Gender'])
```

```python
data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder=OneHotEncoder()
geo_encoder=onehot_encoder.fit_transform(data[['Geography']]).toarray()
```

```python
geo_encoder
```

```python
onehot_encoder.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df=pd.DataFrame(geo_encoder,columns=onehot_encoder.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
data=pd.concat([data.drop(['Geography'],axis=1),geo_encoded_df],axis=1)
```

```python
data.head()
```

```python
with open('labelencoder.pkl','wb') as file:
    pickle.dump(labelencoder,file)

with open('onehot_encoder.pkl','wb') as file:
    pickle.dump(onehot_encoder,file)
```

```python
X=data.drop('Exited',axis=1)
```

```python
y=data['Exited']
```

```python
X.head()
```

```python
X=X.drop('Surname',axis=1)
```

```python
y.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=600)
```

```python
Scaler=StandardScaler()
```

```python
X_train=Scaler.fit_transform(X_train)
X_test=Scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
with open('Scaler.pkl','wb')as file:
    pickle.dump(Scaler,file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
import datetime
```

```python

model=Sequential([
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
```

```python

model.compile(optimizer=opt,loss=loss,metrics=['accuracy'])
```

```python
##Set up the tensorBoard
log_dir="logs/fit/"+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
```

```python
from tensorflow.keras.callbacks import TensorBoard,EarlyStopping
tensorflow_callbacks=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
##set up EarlyStopping
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
history=model.fit(
    X_train,y_train,validation_data=(X_test,y_test),epochs=100,
    callbacks=[tensorflow_callbacks,early_stopping_callback]

)
```

```python
model.save('model.h5')
```
