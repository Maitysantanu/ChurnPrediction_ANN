```python
import tensorflow as tf
import pandas as pd
import numpy as np
import pickle
from tensorflow.keras.models import load_model
```

```python
##Load the trained model,scaler and encoder pickel
model=load_model('model.h5')

with open('labelencoder.pkl','rb') as file:
    label_encoder=pickle.load(file)

with open('onehot_encoder.pkl','rb') as file:
    onehot_encoder=pickle.load(file)

with open('Scaler.pkl','rb') as file:
    scaler=pickle.load(file)

```

```python
input_data={
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000
}
```

```python
geo_encoded=onehot_encoder.transform([[input_data['Geography']]]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoded,columns=onehot_encoder.get_feature_names_out(['Geography']))
```

```python
input_df=pd.DataFrame([input_data])
```

```python
input_df.drop(columns=['Geography'],inplace=True)
```

```python
input_data=pd.concat([input_df,geo_encoded_df],axis=1)
```

```python
input_df['Gender']=label_encoder.transform(input_df['Gender'])
```

```python
input_df_scaled=scaler.transform(input_df)
```

```python
prediction=model.predict(input_df_scaled)
```

```python
prediction_prob=prediction[0][0]
```
